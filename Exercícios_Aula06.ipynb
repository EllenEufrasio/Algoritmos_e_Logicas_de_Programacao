posto de atendimento determinou que em todos os atendimentos o médico ou o
enfermeiro deverão calcular o peso ideal de todos os pacientes atendidos.
Elabore um programa que verifique se o paciente está acima de seu peso ideal
de acordo com as condições abaixo:

• Para homens: (72,7 altura) – 58;

• Para mulheres: (62,1 altura) – 44,7

Para o exercício 06 considerar a fórmula para se obter o peso ideal:
Masculino ==> PesoIdeal = 52 + (0.75 × (altura - 152.4)
Feminino ==>  PesoIdeal = 52 + (0.67 × (altura - 152.4)


```python
print("\n--------- CÁLCULO - PESO IDEAL ---------\n")
peso = float(input("Digite o peso do(a) paciente em quilogramas: "))
alt = float(input("Digite a altura do(a) paciente em centímetros: "))
print("\n")

print("---------------- GENERO ----------------\n")
print("F - FEMININO")
print("M - MASCULINO")
print("----------------------------------------\n")
genero = input("Digite uma das opções acima: ")
print("\n")
if genero in ["feminino", "f","Feminino","FEMININO", "F"]:
    peso_ideal = 52 + (0.67 * (alt - 152.4))
elif genero in ["masculino", "m", "Masculino", "MASCULINO","M"]:
    peso_ideal = 52 + (0.75 * (alt - 152.4))

if peso > peso_ideal:
 print("O(A) paciente está acima do peso ideal.\nO peso ideal é ", peso_ideal)
elif peso < peso_ideal:
 print("O(A) paciente está abaixo do peso ideal.\nO peso ideal é ", peso_ideal)
elif peso == peso_ideal:
 print("O(A) paciente está com peso ideal.")

```
